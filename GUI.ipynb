```python
import tkinter as tk
import subprocess
import os
import time
```

```python


class MyApplication(tk.Tk):
        def __init__(self):
            super().__init__()
            self.title("Or-Q")
            self.geometry("600x400")
            self.shared_data = tk.StringVar(value = "  ") # Upper text
            self.text_data =  tk.StringVar(value = "  ")
            self.ifstart_OMNI = tk.BooleanVar(value = False)
            self.ifstart_IMED = tk.BooleanVar(value = False)
            self.omni_button_color = tk.StringVar(value = "grey")
            self.name_var = tk.StringVar()
            self.entry = tk.Entry(self, textvariable=self.name_var, width=40)
            self.label1 = tk.Label(self, textvariable=self.shared_data)
            self.label1.pack(pady=10)
            self.label2 = tk.Label(self, textvariable=self.text_data)
            self.label2.pack(pady=10)
            self.button = tk.Button(self, text="Start omnibus", bg= self.omni_button_color.get(), command=self.command_called)

            self.text_box = tk.Text(self, height=3, width=50) # Text box
            self.text_box.pack(pady=5)
            
            self.quit_button = tk.Button(self, text='Quit', width=10, command=self.destroy)
            self.button.pack(pady=5)
            self.quit_button.pack(pady=5)

        def command_called(self):
    
            returncode = subprocess.run(["python", r"C:\Users\user\Desktop\OMNI\TRY.py", "It is started"], capture_output=True, text=True)
            self.shared_data.set(returncode.stdout.strip()) # Upper text
            
            self.name_var.get()
            self.ifstart_OMNI.set(True)
            self.omni_button_color.set("green")
            self.button.config(bg=self.omni_button_color.get())
            text = self.text_box.get("1.0", tk.END).strip() # Text box
            
            self.text_data.set(text)
            self.label2.config(text=text) # Lower text
            


if __name__ == "__main__":
    app = MyApplication()
    app.mainloop()
    
```

```python
def folder_checker(folder_path):
    i=0
    dict = {}
    dict_LEN = 0 
    while i < 5:
        j = 0
        with os.scandir(folder_path) as entries:
            for entry in entries:
                if not entry.is_dir():
                    j += 1
    
            if j != len(dict):
                dict = {}
    
        with os.scandir(folder_path) as entries:
            for i,entry in enumerate(entries):
                if not entry.is_dir():
                    if entry.name not in dict:
                        dict[entry.name] = i
                        dict_LEN = len(dict)
                        
        print(dict_LEN)
                    
        i+=1
        time.sleep(1)
```
